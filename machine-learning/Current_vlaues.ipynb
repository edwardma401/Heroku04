```python
import requests
import json
import pandas as pd
from pprint import pprint
from datetime import datetime
import wbdata 
```

```python
df = pd.read_csv("covid_data.csv")
```

```python
df = df.drop_duplicates(subset=['ISO'])
df.head()
```

```python
df = df[["country", "ISO2", "ISO","slug" ]]
```

```python
base_url = "https://api.covid19api.com/total/country/"

start_date = []
end_date = []
active= []
new_cases_feature = []

for i in range(len(df)):
   
    print(df.iloc[i, 3])
    
    request = requests.get(base_url +  df.iloc[i, 3])
    json = request.json()
    json = json[-20:]
        
    confirmed = json[19]["Confirmed"]
    confirmed_2 = json[0]["Confirmed"]

    new_cases_feature_1 = confirmed - confirmed_2

    active_1 = json[19]["Active"]
    end_date_1 = json[19]["Date"]
    start_date_1 = json[0]["Date"]


    start_date.append(start_date_1)
    end_date.append(end_date_1) 
    active.append(active_1)
    new_cases_feature.append(new_cases_feature_1)
    
        

   
        
```

```python
df["start_date"] = start_date
df["end_date"] = end_date
df["active"] = active
df["new_cases_feature"] = new_cases_feature
```

```python
df["end_date"] = pd.to_datetime(df["end_date"]).dt.strftime('%Y-%m-%d')
df["start_date"] = pd.to_datetime(df["start_date"]).dt.strftime('%Y-%m-%d')
```

```python
today = datetime.today()
today = str(today.year) + '-' + str(today.month) + '-' + str(today.day-1)
```

```python
df.head()
```

```python
url = "https://covidtrackerapi.bsg.ox.ac.uk/api/v2/stringency/date-range/" + today + "/" + today
jsonfile_stringency = requests.get(url)
stringency = jsonfile_stringency.json()
```

```python
stringency_actual = []


for i in range(len(df)):

    print(df.iloc[i,2])
    
    stri = stringency["data"][df.iloc[i,5]][df.iloc[i,2]]["stringency"]
    stringency_actual.append(stri)
```

```python
df["stringency_actual"] = stringency_actual
```

```python
countries = list(df["ISO"].values)
```

```python
countries
```

```python
date = datetime(2019, 1, 1)




population = wbdata.get_data("SP.POP.TOTL", country = countries, data_date = date,freq='Y')
gdp_cap = wbdata.get_data("NY.GDP.PCAP.CD", country = countries, data_date = date,freq='Y')
population_density = wbdata.get_data("EN.POP.DNST", country = countries, data_date = datetime(2018, 1, 1),freq='Y')
Urban_population = wbdata.get_data("SP.URB.TOTL.IN.ZS", country = countries, data_date = date,freq='Y')
UHC_service_coverage_index = wbdata.get_data("SH.UHC.SRVS.CV.XD", country = countries, data_date = datetime(2017, 1, 1),freq='Y')
basic_sanitation_services = wbdata.get_data("SH.STA.BASS.ZS", country = countries, data_date = datetime(2017, 1, 1),freq='Y')
Population_ages_0_to_14 = wbdata.get_data("SP.POP.0014.TO.ZS", country = countries, data_date = date,freq='Y')
Population_ages_15_to_64 = wbdata.get_data("SP.POP.1564.TO.ZS", country = countries, data_date = date,freq='Y')
Population_ages_65_and_over = wbdata.get_data("SP.POP.65UP.TO.ZS", country = countries, data_date = date,freq='Y')


```

```python
multiple_data = {}

for i in range(len(Urban_population)):
    multiple_data.update({Urban_population[i]["countryiso3code"] : {"population" : population[i]["value"],
                                                                "gdp_per_capita": gdp_cap[i]["value"],
                                                                "population_density" : population_density[i]["value"],
                                                                "Urban_population" : Urban_population [i]["value"], 
                                                                 "UHC_service_coverage_index": UHC_service_coverage_index [i]["value"],
                                                                "basic_sanitation_services": basic_sanitation_services [i]["value"],
                                                                 "Population_ages_0_to_14": Population_ages_0_to_14 [i]["value"],
                                                                  "Population_ages_15_to_64": Population_ages_15_to_64 [i]["value"],
                                                                   "Population_ages_65_and_over": Population_ages_65_and_over [i]["value"]}})
```

```python
population_list = []
gdp_per_capita_list = []
population_density_list = []
Urban_population_list = []
UHC_service_coverage_index_list = []
basic_sanitation_services_list = []
Population_ages_0_to_14_list = []
Population_ages_15_to_64_list = []
Population_ages_65_and_over_list = []

for i in range(len(df)):

    d1 = multiple_data[df.iloc[i,2]]["Urban_population"]
    d2 = multiple_data[df.iloc[i,2]]["UHC_service_coverage_index"]
    d3 = multiple_data[df.iloc[i,2]]["basic_sanitation_services"]
    d4 = multiple_data[df.iloc[i,2]]["Population_ages_0_to_14"]
    d5 = multiple_data[df.iloc[i,2]]["Population_ages_15_to_64"]
    d6 = multiple_data[df.iloc[i,2]]["Population_ages_65_and_over"]
    d7 = multiple_data[df.iloc[i,2]]["population"]
    d8 = multiple_data[df.iloc[i,2]]["gdp_per_capita"]
    d9 = multiple_data[df.iloc[i,2]]["population_density"]
    
    population_list.append(d7)
    gdp_per_capita_list.append(d8)
    population_density_list.append(d9)
    Urban_population_list.append(d1)
    UHC_service_coverage_index_list.append(d2)
    basic_sanitation_services_list.append(d3)
    Population_ages_0_to_14_list.append(d4)
    Population_ages_15_to_64_list.append(d5)
    Population_ages_65_and_over_list.append(d6)
    
```

```python
df["population"] = population_list
df["gdp_per_capita"] = gdp_per_capita_list
df["population_density"] = population_density_list

df["Urban_population"] = Urban_population_list
df["UHC_service_coverage_index"] = UHC_service_coverage_index_list
df["basic_sanitation_services"] = basic_sanitation_services_list
df["Population_ages_0_to_14"] = Population_ages_0_to_14_list
df["Population_ages_15_to_64"] = Population_ages_15_to_64_list
df["Population_ages_65_and_over"] = Population_ages_65_and_over_list
```
